# Exploring Hacker News Posts

Hacker News is a site started by the startup incubator Y Combinator, where user-submitted stories (known as "posts") receive votes and comments, similar to reddit. Hacker News is extremely popular in technology and startup circles, and posts that make it to the top of the Hacker News listings can get hundreds of thousands of visitors as a result. You can find the data set [here](https://www.kaggle.com/datasets/hacker-news/hacker-news-posts)

* We're specifically interested in posts with titles that begin with either `Ask HN` or `Show HN`. Users submit `Ask HN` posts to ask the Hacker News community a specific question.

* Likewise, users submit `Show HN` posts to show the Hacker News community a project, product, or just something interesting.

We'll compare these two types of posts to determine the following:

* Do `Ask HN` or `Show HN` receive more comments on average?
* Do posts created at a certain time receive more comments on average?


# Introduction

```python
from csv import reader
with open("hacker_news.csv") as opened_file:
    read_file = reader(opened_file)
    hn = list(read_file)
    
print(hn[:5])
```

#  Removing Headers from a List of Lists

```python
headers = hn[0]
hn = hn[1:]
print(headers)
print(hn[:5])
```

# Extracting Ask HN and Show HN Posts

```python
ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row [1]
    if title.lower().startswith('ask hn'):
        ask_posts.append(row)
    elif title.lower().startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)
        
print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))
```

# Calculating the Average Number of Comments for Ask HN and Show HN Posts

```python
total_ask_comments = 0

for row in ask_posts:
    total_ask_comments += int(row[4])

avg_ask_comments = total_ask_comments / len(ask_posts)
print(avg_ask_comments)
```

```python
total_show_comments = 0

for row in show_posts:
    total_show_comments += int(row[4])

avg_show_comments = total_show_comments / len(show_posts)
print(avg_show_comments)
```

The amount of comments ask_posts gets on average is 14 and for show_posts its 10. Since ask_posts gets more comments we will focus on that.

# Finding the Number of Ask Posts and Comments by Hour Created

```python
import datetime as dt

result_list = []

for row in ask_posts:
    created_at = row[6]
    result_list.append([created_at, int(row[4])])
    
counts_by_hour = {}
comments_by_hour = {}
date_format = "%m/%d/%Y %H:%M"

for row in result_list:
    date = row[0]
    comment = row[1]
    time = dt.datetime.strptime(date, date_format).strftime("%H")
    if time not in counts_by_hour:
        counts_by_hour[time] = 1
        comments_by_hour[time] = comment
    else:
        counts_by_hour[time] += 1
        comments_by_hour[time] += comment 
```

#  Calculating the Average Number of Comments for Ask HN Posts by Hour

```python
avg_by_hour = []

for hour in comments_by_hour:
    avg_by_hour.append([hour, comments_by_hour[hour] / counts_by_hour[hour]])

print(avg_by_hour)
```

# Sorting and Printing Values from a List of Lists

```python
swap_avg_by_hour = []

for row in avg_by_hour:
    swap_avg_by_hour.append([row[1], row[0]])
    
swap_avg_by_hour
```

```python
sorted_swap = sorted(swap_avg_by_hour, reverse = True)
sorted_swap
```

```python
print("Top 5 Hours for Ask Posts Comments")

for avg,hour in sorted_swap[:5]:
    print(f"{dt.datetime.strptime(hour, '%H').strftime('%H:%M')} : {avg:.2f} average comments per post")
```

After analyzing the result above it is clear that the hour 15:00 which can also be 3 p.m. has the highest average comments. Note: The time in this dataset is in EDT (UTC-4) time zone.

# Conclusion

We analyzed both ask posts and show posts in this project to find out which type of posts gets the most comments by hour.

After the analysis we determined that ask posts at 15:00 EDT has the highest chance of getting comments which amounts to 38.59. So, we recommend that in order to get the most comments on a post it must be a ask post and it should be in between 15:00 and 16:00.
